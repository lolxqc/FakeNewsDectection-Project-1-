```python
data_fake_manual_testing=data_fake_manual_testing.copy()
data_true_manual_testing=data_true_manual_testing.copy()
data_fake_manual_testing.loc[:,'class']=0
data_true_manual_testing.loc[:,'class']=1
```

```python
data_merge=pd.concat([data_fake,data_true],axis=0)
```

```python
data.reset_index(inplace=True)
data.drop(['index'],axis=1,inplace=True)
```

```python
import re
import string
def wordopt(text):
    text = text.lower()
    text = re.sub(r'\[.*?\]', '', text)
    text = re.sub(r"\W", " ", text)
    text = re.sub(r'https?://\S+', '', text)  # fixed \S+
    text = re.sub(r'<.*?>+', '', text)
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub(r'\n', '', text)  # fixed /n to \n and made raw
    text = re.sub(r'\w*\d\w*', '', text)
    return text
```

```python
data['text']=data['text'].apply(wordopt)
x=data['text']
y=data['class']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,stratify=y,random_state=0)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer(
    max_df=0.7,           # Ignore terms in more than 70% of documents
    min_df=5,             # Ignore rare terms in fewer than 5 documents
    stop_words='english', # Remove common stopwords
    max_features=5000     # Use only top 5000 features (to reduce noise)
)

xv_train=vectorizer.fit_transform(x_train)
xv_test=vectorizer.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression(class_weight='balanced')
LR.fit(xv_train,y_train)
```

```python
pred_lr=LR.predict(xv_test)
```

```python
LR.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier
DT=DecisionTreeClassifier(max_depth=20, min_samples_split=10, class_weight='balanced')
DT.fit(xv_train,y_train)
```

```python
pred_dt=DT.predict(xv_test)
```

```python
DT.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_dt))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
GB=GradientBoostingClassifier(random_state=0)
GB.fit(xv_train,y_train)
```

```python
pred_gb=GB.predict(xv_test)
```

```python
GB.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_gb))
```

```python
from sklearn.ensemble import RandomForestClassifier
RF=RandomForestClassifier(n_estimators=50,max_depth=20,random_state=0,n_jobs=-1)
RF.fit(xv_train,y_train)
```

```python
pred_rf=RF.predict(xv_test)
```

```python
RF.score(xv_test,y_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, pred_rf))
```

```python
print("\nModel Accuracy Scores:")
print("Logistic Regression:", LR.score(xv_test, y_test))
print("Decision Tree:", DT.score(xv_test, y_test))
print("Gradient Boosting:", GB.score(xv_test, y_test))
print("Random Forest:", RF.score(xv_test, y_test))
```

```python
def output_label(n):
    return "true" if n == 1 else "fake"
from collections import Counter
def ensemble_prediction(preds):
    return Counter(preds).most_common(1)[0][0]
def manual_testing(news):
    testing_news = {"text": [news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)
    new_xv_test = vectorizer.transform(new_def_test["text"])

    pred_LR = LR.predict(new_xv_test)[0]
    pred_DT = DT.predict(new_xv_test)[0]
    pred_GB = GB.predict(new_xv_test)[0]
    pred_RF = RF.predict(new_xv_test)[0]

    predictions = [pred_LR, pred_DT, pred_GB, pred_RF]
    final_vote = ensemble_prediction(predictions)

    print(f"\nNews: {news}")
    print("Logistic Regression Prediction:", output_label(pred_LR))
    print("Decision Tree Prediction:", output_label(pred_DT))
    print("Gradient Boosting Prediction:", output_label(pred_GB))
    print("Random Forest Prediction:", output_label(pred_RF))
    print("🗳️ Final Ensemble Vote:", output_label(final_vote))
print("\n🔍 Testing known REAL samples (from True.csv manual set):")
for i in range(5):
    news = data_true_manual_testing.iloc[i]['text']
    manual_testing(news)
print("\n🔍 Testing known FAKE samples (from Fake.csv manual set):")
for i in range(5):
    news = data_fake_manual_testing.iloc[i]['text']
    manual_testing(news)
```
